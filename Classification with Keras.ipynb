```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))

model.add(Dense(15,activation='relu'))
#binary classification the last layer has sigmoid
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y = y_train, epochs=600, validation_data = (X_test,y_test),verbose=1 )
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

This particular model has overfitted so the possible two solution - earlystopping

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
#binary classification the last layer has sigmoid
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x= X_train, y=y_train,
         validation_data=(X_test,y_test),
         epochs=600,
         verbose=1,
         callbacks=[earlystop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Next part is adding a dropout layer

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,
         validation_data=(X_test,y_test),
         epochs=600,
         verbose=1,
         callbacks=[earlystop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Model evaluation

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
